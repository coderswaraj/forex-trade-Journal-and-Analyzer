```python
import csv
import os
```

```python
file_name = "trades.csv"

# create file if not exists
if not os.path.exists(file_name):
    f = open(file_name, "w")
    writer = csv.writer(f)
    writer.writerow(["Pair", "Entry", "Exit", "PnL"])
    f.close()
```

```python
print("Welcome to our Forex Market Journal andAnalyzer ")
while True:
    print("1. Add Trade")
    print("2. View Trades")
    print("3. Analyze Trades")
    print("4. Exit")

    choice = input("Enter choice: ")

# adding choice 1 = addtrade
    if choice == "1":
        pair = input("Enter pair (Ex - EURUSD): ")
        entry = float(input("Enter entry price: "))
        exit_price = float(input("Enter exit price: "))

# profit and loss calculation
        
        pnl = exit_price - entry

# giving sequence to data
        
        f = open(file_name, "a")
        writer = csv.writer(f)
        writer.writerow([pair, entry, exit_price, pnl])
        f.close()

        print("Trade saved")

 # giving choice 2 = view trades
    
    elif choice == "2":
        if not os.path.exists(file_name):
            print("No file found")
        else:
            f = open(file_name, "r")
            reader = csv.reader(f)

            for row in reader:
                print(row)

            f.close()

# giving choice no 3 = analyze trades
    
    elif choice == "3":
        if not os.path.exists(file_name):
            print("No data")
        else:
            f = open(file_name, "r")
            reader = csv.DictReader(f)

            total = 0
            wins = 0
            total_pnl = 0

            for row in reader:
                pnl = float(row["PnL"])
                total = total + 1
                total_pnl = total_pnl + pnl

                if pnl > 0:
                    wins = wins + 1

            f.close()

            if total == 0:
                print("No trades")
            else:
                losses = total - wins
                win_rate = (wins / total) * 100

                print("Total trades:", total)
                print("Wins:", wins)
                print("Losses:", losses)
                print("Win rate:", win_rate)
                print("Total pnl:", total_pnl)

    # EXIT
    elif choice == "4":
        print("Thank you!")
        break

    else:
        print("Wrong choice")
```
